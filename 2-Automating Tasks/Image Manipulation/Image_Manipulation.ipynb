Let's start by importing the modules needed for image manipulation.

```python
from PIL import Image, ImageFilter
import os
```

## Creating Image Objects
Lets begin by opening an image.

```python
image1 = Image.open('bear1.jpg')  # Create image obj
image1
```

## Iteration
Soooo cute! Let's iterate to collect all bear images and convert them to png.

```python
for f in os.listdir('.'):
    if f.endswith('.jpg'):
        i = Image.open(f)
        fn, fext = os.path.splitext(f)
        print(fn)
        i.save('pngs/{}.png'.format(fn))  # save as png in pngs folder
```

## Resizing files
Lets resize these cute bears so theres more of them to love.

```python
size_300 = (300,300)  # size must be a tuple
size_700 = (700,700)

for f in os.listdir('.'):
    if f.endswith('.jpg'):
        i = Image.open(f)
        fn, fext = os.path.splitext(f)

        i.thumbnail(size_700)
        i.save('700/{}_700{}'.format(fn, fext))  # save as 300x300ng in pngs folder

        i.thumbnail(size_300)
        i.save('300/{}_300{}'.format(fn, fext))  # save as 300x300png in pngs folder
```

## More Manipulation
### Rotation
First, lets rotate a bear. To rotate images, use `.rotate`

```python
image1 = Image.open('bear1.jpg')
image1 = image1.rotate(90)
image1.save('bear1_mod.jpg')  # save the sideways bear
image1
```

### Black & White
Gorgeous. Now, lets make the bear black and white. To convert from RGB to monochrome, use `.convert`

```python
image1 = Image.open('bear1_mod.jpg')
image1 = image1.convert('L')
image1.save('bear1_mod.jpg')
image1
```

### Image Blur
Grizzly is looking a little spooky, so let's blur him. To blur images, use `.ImageFilter`

```python
image1 = Image.open('bear1_mod.jpg')
image1 = image1.filter(ImageFilter.GaussianBlur(2))  # blur values are radius set to 2
image1.save('bear1_mod.jpg')
image1
```
